```python
import numpy as np

def simulate_urn():
    urn = {"red":3, "blue":4, "black":2}

    die_roll = np.random.randint(1,7)

    if die_roll in [2, 3, 5]: #check by die face
        urn["black"] += 1
        color = "black"
    elif die_roll == 6:
        urn["red"] += 1
        color = "red"
    else:
        urn["blue"] += 1
        color = "blue"

    return color
```

```python
nr_of_trials = 10000
```

```python
red_rolled = 0

for _ in range(nr_of_trials):
    red_rolled += 1 if simulate_urn()=="red" else 0

print(f"Estimated probability to get a red ball: {red_rolled/nr_of_trials:.2f}")
```

```python
p_is6 = 1/6
p_is_prime = 3/6
p_other = 2/6

p_red_is6 = 4/10
p_red_is_prime = 3/10
p_red_other = 3/10

p_red = p_red_is6 * p_is6 + p_red_is_prime * p_is_prime + p_red_other * p_other

print(f"Estimated probability to get a red ball: {red_rolled/nr_of_trials:.2f}")
print(f"Theoretical probability to get a red ball: {p_red:.2f}")
```

```python
import numpy as np


np.random.seed(123)
```

```python
lambdas = [1, 2, 5, 10]
simulations_nr = 1000

simulation_results = {}
# Simulation of 1000 values for each lambds
for lam in lambdas:
    key = f"Poison({lam})"
    simulation_results[key] = np.random.poisson(lam=lam, size=simulations_nr)
# Output
for lam in lambdas:
    key = f"Poison({lam})"
    data = simulation_results[key]
    print(f"{key}")
    print(f"Mean={data.mean()}")
    print(f"Variance={data.var()}")
    print(f"First 10 values: {data[:10]}")
```
