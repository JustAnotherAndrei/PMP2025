```python
# bayesian_noise_analysis.py
#am folosit priori mari, cu media esantionului ca centru pentru priorul lui μ
#am obtinut intervalele HDI de 95% pentru μ si σ, care indica unde se afla cei mai probabili 95% din valorile posteriorului
#in modelul cu prior puternic, am dus posteriorul spre 50, ceea ce arata influenta priorului asupra rezultatului

import numpy as np
import pymc as pm
import arviz as az

observatii = np.array([56, 60, 58, 55, 57, 59, 61, 56, 58, 60])
x = np.mean(observatii)

# model bayesian cu prior slab
with pm.Model() as model:
    mu = pm.Normal("mu", mu=x, sigma=10)
    sigma = pm.HalfNormal("sigma", sigma=10)
    zgomot = pm.Normal("zgomot", mu=mu, sigma=sigma, observed=observatii)
    trace = pm.sample(2000, tune=1000, return_inferencedata=True)

# interval HDI de 95%
print("\n--- HDI 95% pentru prior slab ---")
hdi = az.hdi(trace, hdi_prob=0.95)
print(hdi)

# estimari frecventiste
media_frecventista = np.mean(observatii)
dev_std_frecventista = np.std(observatii, ddof=1)

print("\n--- Estimari frecventiste ---")
print(f"Media frecventista: {media_frecventista:.2f}")
print(f"Deviatia standard frecventista: {dev_std_frecventista:.2f}")

# model bayesian cu prior puternic
with pm.Model() as model_strong_prior:
    mu_strong = pm.Normal("mu", mu=50, sigma=1)
    sigma_strong = pm.HalfNormal("sigma", sigma=10)
    zgomot_strong = pm.Normal("zgomot", mu=mu_strong, sigma=sigma_strong, observed=observatii)
    trace_strong = pm.sample(2000, tune=1000, return_inferencedata=True)

print("\n--- HDI 95% pentru prior puternic ---")
hdi_strong = az.hdi(trace_strong, hdi_prob=0.95)
print(hdi_strong)

```
